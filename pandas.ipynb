# What is Pandas?

Pandas is a Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.

The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

# Why Use Pandas?

Pandas allows us to analyze big data and make conclusions based on statistical theories.

Pandas can clean messy data sets, and make them readable and relevant.

Relevant data is very important in data science.

```python
import pandas as pd
import numpy as np
```

# Basic data structures in pandas
Pandas provides two types of classes for handling data:
1. Series: a one-dimensional labeled array holding data of any type
2. DataFrame: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

```python
arr1=pd.Series([1,2,3])
arr1
```

```python
type(arr1)
```

```python
arr2=pd.Series([1,2,3],index=['a','b','c'])
arr2
```

```python
df1=pd.DataFrame({    
    "B": [1,2,3],
    "C": ["Ayan","Dim","Nahid"],
    "D": [1.5,2.5,3.5],
    "E": [0,1,2]})
df1
```

```python
# Correlation of all numeric columns with 'Survived'
corr_with_survived = data.corr(numeric_only=True)['Survived'].sort_values(ascending=False)
```

```python
corr_with_survived 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Correlation matrix for numeric columns
corr_matrix = data.corr(numeric_only=True)

# Display as heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix (Numeric Features)")
plt.show()
```
